CLUSTERING MODELS 

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans 
```

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=4,cluster_std=1,
               center_box=(-10.0,10.0),
              shuffle=True,random_state=1)
range_n_clusters=[2,3,4,5,6]

#samples=500-total no of data points
#n_features=2 - each point has 2 coords 
#centers=4 generate 4 distinct clusters
#cluster_std=1 how spread out each clister must be
#center_box= the cluster is randonmly placed b/w this coord range
#shuffle=True  shuffles the order of points 
#random_state=1 same result eberytime code runs
```

```python
y
```

```python
X
```

```python
#elbow method-used to find optimised k value 
wcss=[] 
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
# within cluster sum of squares
# looping from 1-10 running i clusters and using kmeans++ to get initial proper
# placing of centroids
#kmeans.intertia_= lower inertia tighter clusters 

```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# here last abrupt change is 4  hence select k=4
```

```python
#Now using silhoutte v r going to validate if k=4 is better 
```

```python
from sklearn.metrics import silhouette_score
```

```python
for n_clusters in range_n_clusters:
    cluster=KMeans(n_clusters=n_clusters,random_state=10)
    cluster_labels=cluster.fit_predict(X) #training 
    sil_avg=silhouette_score(X,cluster_labels) # trained and tested 
    print("for",n_clusters, "clusters the avg score is :", sil_avg)
```

```python
sil_avg=silhouette_score(X,cluster_labels) # trained and tested 
print("for 4 clusters the avg score is :", sil_avg)
```
